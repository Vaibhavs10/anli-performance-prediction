```python
import json
import csv
```

```python
story_id = []
obs1 = []
obs2 = []
hyp1 = []
hyp2 = []

with open('data/train.jsonl') as json_file:
    for line in json_file:
        dict_line = json.loads(line)
        story_id.append(dict_line["story_id"])
        obs1.append(dict_line["obs1"])
        obs2.append(dict_line["obs2"])
        hyp1.append(dict_line["hyp1"])
        hyp2.append(dict_line["hyp2"])        
```

```python
train_labels = []
with open('data/train-labels.lst') as file:
    for line in file:
        if line != " ":
            train_labels.append(line.replace("\n", ""))
```

```python
rows = zip(story_id,obs1,obs2,hyp1,hyp2,train_labels)

with open("train.csv", "w") as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
